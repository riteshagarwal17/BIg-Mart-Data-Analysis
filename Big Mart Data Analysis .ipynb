# EXPLORATORY DATA ANALYSIS OF BIG MART DATA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv("Big Mart Data.csv")
```

```python
df1.isnull().sum()
```

```python
df=df1
df.head()
```

```python
df['Item_Weight']=df['Item_Weight'].fillna(df['Item_Weight'].mean())
df['Item_Weight'].isnull().sum()
```

```python
df['Outlet_Size']=df['Outlet_Size'].fillna(df['Outlet_Size'].mode()[0])
df['Outlet_Size'].isnull().sum()
df['Outlet_Size'][df['Item_Identifier']=="FDX07"]
```

```python
df["Item_Type"].unique()
```

```python
df["Item_Fat_Content"].unique()
```

```python
df["Outlet_Type"].unique()
```

```python
print(df[df["Outlet_Type"]=="Supermarket Type3"]["Outlet_Type"].value_counts())
print(df[df["Outlet_Type"]=="Supermarket Type2"]["Outlet_Type"].value_counts())
print(df[df["Outlet_Type"]=="Supermarket Type1"]["Outlet_Type"].value_counts())
print(df[df["Outlet_Type"]=="Grocery Store"]["Outlet_Type"].value_counts())
```

```python
df["Item_Fat_Content"]=df["Item_Fat_Content"].replace(['low fat','LF','reg'],['Low Fat','Low Fat','Regular'])
df["Item_Fat_Content"].unique()
```

```python
#print(df[df["Item_Visibility"]==0.000000]["Item_Visibility"].value_counts())
df['Item_Visibility']=df['Item_Visibility'].replace([0.000000],[df["Item_Visibility"].mean()])
df[df["Item_Visibility"]==0.000000]["Item_Visibility"].count()
```

```python
#plt.grid()
plt.title("Item_Visibility COUNTPLOT")
sns.distplot(df['Item_Visibility'],color='Green',hist=True)
sns.despine()
```

```python
sns.regplot(x="Item_Visibility",y="Item_Outlet_Sales",x_jitter=0.3,data=df,scatter_kws={"color": "purple"}, line_kws={"color": "red"})
sns.despine()
plt.grid()
plt.title("Item_Visibility Vs Item_Outlet_sales Plot")
```

```python
ax=sns.boxplot(x="Outlet_Type",y="Item_Outlet_Sales",data=df)
ax.set_xticklabels(ax.get_xticklabels(),rotation=10)
plt.grid()
```
